# Images as Grid of Pixels

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_dir = 'images/waymo_car.jpg'

#Read image
image = cv2.imread(image_dir)

# Print out the image dimensions
print('Image dimensions:', image.shape)

image2 = mpimg.imread(image_dir)

fig,(ax1,ax2) = plt.subplots(1,2,figsize = (10,8))
ax1.matshow(image)
ax1.set_title('OpenCV')
ax2.matshow(image2)
ax2.set_title('Matplotlib')
```

There is difference in above 2 images because OpenCV reads the 3 colour channels in order 'BGR' while matplotlib reads the 3 channels in order 'RGB'.

```python
#convert BGR to RGB using OpenCV
convert_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.matshow(convert_img)
```

```python
#Grayscale Image
gray = cv2.cvtColor(convert_img,cv2.COLOR_RGB2GRAY)

plt.matshow(gray,cmap='gray')

print('Shape of original Image: ',convert_img.shape)
print('Shape of GrayScaled Image: ',gray.shape)
```

```python
#Find the maximum and minimum values in the grayscaled image

max_val = np.amax(gray)
min_val = np.amin(gray)

print('Max: ', max_val)
print('Min: ', min_val)
```

### Some tips regarding Reading of Images using different libraries

* When we use OpenCV to read images then the pixel values are 0-255
* Matplotlib natively supports only 'PNG' format of the images and pixel values are normalized between   0-1
* Matplotlib uses PIL in background to read images which are in format other than 'PNG' and pixel    values are 0-255

```python
img=cv2.imread('images/waymo_car.jpg') #0-255
img=cv2.imread('images/waymo_car.png') #0-255
img=plt.imread('images/waymo_car.png') #0-1
img=plt.imread('images/waymo_car.jpg') #0-255
img=mpimg.imread('images/waymo_car.jpg') #0-255
img=mpimg.imread('images/waymo_car.png') #0-1
```
